### Silver Layer - One Off Taxi Zone Lookup Table Creation

This notebook creates one off taxi zone look up table from the data file ingested for this purpose. It will be used for slowly changing dimensions(SCD) type 2.

```python
# import statments
from pyspark.sql.functions import col, current_timestamp, lit
from pyspark.sql.types import IntegerType, TimestampType
```

```python
# loading the file
df = spark.read.format("csv").option("header", True).load("/Volumes/nyctaxi/00_landing/data_sources/taxi_zone_lookup")
```

```python
# creating the dataframe with columns effective_date and end_date for SCD type 2
df = df.select(
    col("LocationID").cast(IntegerType()).alias("location_id"),
    col("Borough").alias("borough"),
    col("Zone").alias("zone"),
    col("service_zone"),
    current_timestamp().alias("effective_date"),
    lit(None).cast(TimestampType()).alias("end_date")
)
```

```python
# saving the table
df.write.mode("overwrite").saveAsTable("nyctaxi.02_silver.taxi_zone_lookup")
```
