### Bronze layer - Yellow Taxi Rides Table 

This notebook reads the newly arrived(Data lag is two months) file and save its contents, after adding column for processed timestamp, as delta table.

```python
# import statements
from datetime import datetime
from pyspark.sql.functions import current_timestamp
from dateutil.relativedelta import relativedelta
```

```python
# calculating two months ago
two_months_ago = (datetime.today() - relativedelta(months=2)).strftime("%Y-%m")
```

```python
# reading the file
df = spark.read.format("parquet").load(f"/Volumes/nyctaxi/00_landing/data_sources/yellow_taxi/{two_months_ago}")
```

```python
# adding processed timestamp
df = df.withColumn("processed_timestamp", current_timestamp())
```

```python
# saving as the table
df.write.mode("overwrite").saveAsTable("nyctaxi.01_bronze.yellow_taxi_trips_bronze")
```
