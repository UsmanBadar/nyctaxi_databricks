### Ingesting Taxi Zone Lookup File 

This notebook fetches two month ago taxi zone data file from the NYC Taxi and Limousine Commission website and stores it.

It has sets the task value (yes or no) to be later used in Lakeflow Job.


```python
# Import statements
import os
from urllib import request
import shutil
```

```python
taxi_zone_file_url = "https://d37ci6vzurychx.cloudfront.net/misc/taxi_zone_lookup.csv"
```

```python
dir_path = f"/Volumes/nyctaxi/00_landing/data_sources/taxi_zone_lookup"
file_path = f"{dir_path}/taxi_zone_lookup.csv"
```

```python
# Setting task value to no
dbutils.jobs.taskValues.set(key="taxi_zone_file_fetched", value = "no")
```

```python
try:
    os.makedirs(dir_path, exist_ok=True)
    response = request.urlopen(taxi_zone_file_url)
    if response.getcode() == 200:
        # Overwrites file if already exists
        with open(file_path, 'wb') as f:
            shutil.copyfileobj(response, f)
        # setting task value to yes on success
        dbutils.jobs.taskValues.set(key="taxi_zone_file_fetched", value = "yes")
except Exception as e:
    print(f"Failed to fetch file {e}")
```
