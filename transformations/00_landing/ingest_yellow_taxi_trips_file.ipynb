### Ingesting Yellow Taxi Trip File 

This notebook fetches two month ago yellow taxi trip raw data from the NYC Taxi and Limousine Commission website and stores it.

This website provides data after a lag of two months. 

It has sets the task value (yes or no) to be later used in Lakeflow Job.

```python
# Import statements
from datetime import datetime
from dateutil.relativedelta import relativedelta
import os
from urllib import request
import shutil
```

```python
# Get current month minus 2
current_month_minus_2 = (datetime.today() - relativedelta(months=2)).strftime("%Y-%m")
```

```python
yellow_taxi_file_url = f"https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_{current_month_minus_2}.parquet"
```

```python
# file storage locaation
dir_path = f"/Volumes/nyctaxi/00_landing/data_sources/yellow_taxi/{current_month_minus_2}"
file_path = f"{dir_path}/yellow_tripdata_{current_month_minus_2}.parquet"
```

```python
# setting the task value
dbutils.jobs.taskValues.set(key="yellow_taxi_file_fetched", value = "no")

```

```python
try:
    os.makedirs(dir_path, exist_ok=True)
    response = request.urlopen(yellow_taxi_file_url)
    if response.getcode() == 200:
        # Overwrites the file if already exists
        with open(file_path, 'wb') as f:
            shutil.copyfileobj(response, f)
        dbutils.jobs.taskValues.set(key="yellow_taxi_file_fetched", value = "yes")
except Exception as e:
    print(f"Failed to fetch file {e}")
```
