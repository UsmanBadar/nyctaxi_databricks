### Ceating Test
This notebook contains tests to validate data.

```python
# Import statement
from pyspark.sql.functions import col
```

```python
# Read the table
df = spark.read.table("nyctaxi.02_silver.yellow_taxi_trips_enriched_silver")
```

```python

# Function to validate enriched trips
def validate_enriched_trips(df):

    tests = {
        "Pickup borough not null":
            df.filter(col("pu_borough").isNull()).count() == 0,

        "Dropoff borough not null":
            df.filter(col("do_borough").isNull()).count() == 0,

        "Trip distance valid":
            df.filter(col("trip_distance") < 0).count() == 0,
        
        "Trip valid":
            df.filter(col("passenger_count") < 0).count() == 0,

        "Pickup before dropoff":
            df.filter(
                col("tpep_dropoff_datetime") <
                col("tpep_pickup_datetime")
            ).count() == 0
    }

    for test, passed in tests.items():
        if passed:
            print(f"PASS: {test}")
        else:
            raise Exception(f"FAILED: {test}")
```

```python
# Calling the function
validate_enriched_trips(df)
```
